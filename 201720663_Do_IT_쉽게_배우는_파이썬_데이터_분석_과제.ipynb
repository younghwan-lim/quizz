```
# Q1
# 데이터 특징을 파악
# head(), tail(), shape, info(), describe()

# 앞에서 5개 데이터 추출하기

midwest.head() # 앞부분 출력
midwest.tail() # 뒷부분 출력
midwest.shape  # 행, 열 개수 출력
midwest.info() # 변수 속성 출력
midwest.describe() # 요약 통계량 출력
```

```
# Q2
# poptotal(전체 인구)변수 -> total
# 이름변경, rename()

midwest = midwest.rename(columns={'poptotal':'total'})
midwest
```

```
# popasian(전체 인구)변수 -> asian
# 이름변경, rename()

midwest = midwest.rename(columns={'popasian':'asian'})
midwest
```

```
# Q3
# total : 전체 인구
# asian : 아시안 인구
# 백분율 : * 100
# 파생변수 열 이름은 rario(비율)

midwest['ratio'] = (midwest['asian'] / midwest['total']) * 100
print(midwest)

# 히스토그램, 변수.plot.hist()
midwest['ratio'].plot.hist()
```

```
# Q4
# 아시아 인구 백분율 : midwest['ratio']
# 전페 평균,.mean()

midwest['ratio'].mean()
```

```
# 평균을 초과
# midwest['ratio'].mean() >

# 하면 -> if구문 -> np.where

# 파생변수, group

# midwest['ratio'].mean() = 0.4872461834357345 변경 가능

import numpy as np

midwest['group'] = np.where(midwest['ratio'] > 0.4872461834357345, 'large', 'small')
midwest
```

```
# Q5
# 'large', 'small'에 해당하는 지역이 얼마나 많은지

# 빈도표
# value_counts()
# 변수명 : count_group


count_group = midwest['group'].value_counts()
count_group
```

```
# 빈도 막대 그래프
# 변수명.plot.bar()
count_group.plot.bar(rot=0)
```

```
# 필요한 모듈 import
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```
# quiz.csv
# quiz.csv 를 이용해
quiz = pd.read_csv('quiz.csv', encoding='cp949')
quiz
```

```
# Q1
# CSV 파일에서 에너지 데이터를 사용
energy_df['Energy'] = quiz['AC_W']
```

```
# 각 월별 에너지 발전량의 합 구하기
energy_df['Month'] = energy_df['Date'].dt.month
monthly_energy = energy_df.groupby('Month')['Energy'].sum()
```

```
# 가장 작은 달과 가장 큰 달 찾기
min_month = monthly_energy.idxmin()
max_month = monthly_energy.idxmax()
```

```
# 해당 월의 데이터 추출
min_month_data = energy_df[energy_df['Month'] == min_month]['Energy']
max_month_data = energy_df[energy_df['Month'] == max_month]['Energy']
```

```
# t-검정 수행
t_stat, p_value = ttest_ind(min_month_data, max_month_data)
```

```
# 결과 출력
print(f"가장 작은 달: {min_month}")
print(f"가장 큰 달: {max_month}")
print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```
# 유의수준 설정
alpha = 0.025
```

```
# P-value와 유의수준을 비교하여 유의한지 확인
if p_value < alpha:
    print("가장 작은 달과 가장 큰 달의 에너지 발전량은 유의한 차이가 있습니다.")
else:
    print("가장 작은 달과 가장 큰 달의 에너지 발전량은 유의한 차이가 없습니다.")
```
